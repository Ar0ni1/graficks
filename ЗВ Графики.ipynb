Шаг 1. Импортируем pandas для работы с датасетом и matplotlib для работы с графиками 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
starWars = pd.read_csv('star_wars_character_dataset.csv')
starWars
```

Шаг 2. Сделаем 3 штуки линейного графика

```python
plt.figure(figsize=(15, 5))
plt. plot(starWars['name'].head(10), starWars['height'].head(10), label='рост')
plt.title('Рост персонажей')
plt.xlabel('Имя персонажа')
plt.ylabel('Рост')
plt.legend()
plt.show()
```

Был сделан линейный график на котором отображен рост первых десяти персонажей (например: рост Люка Скайуокера равен 172, рост C-3PO равен 167 и т.д. )

```python
plt.figure(figsize=(15, 5))
plt. plot(starWars['name'].head(10), starWars['mass'].head(10), label='вес', marker = 'o', markersize = 7)
plt.title('Вес персонажей')
plt.xlabel('Имя персонажа')
plt.ylabel('Вес')
plt.legend()
plt.show()
```

Был сделан линейный график на котором отображен вес первых десяти персонажей (например: вес Люка Скайуокера равен 77, рост C-3PO равен 75 и т.д. ), также на этом графике были добавлены точки на пересечении при помощи параметра marker 

```python
plt.figure(figsize=(15, 5))
plt. plot(starWars['name'].head(10), starWars['birth_year'].head(10), label='возраст', marker = 'o', markersize = 7, color = 'pink')
plt.title('Возраст персонажей')
plt.xlabel('Имя персонажа')
plt.ylabel('Возраст')
plt.legend()
plt.grid()
plt.show()
```

Был создан третий линейный гарфик на котором изображен возраст персонажей и их имя (например: возраст Люка Скайуокера равен 19, возраст C-3PO равен 112) на этом графике видно, что у R5-D4 нету возраста в датасете, это линейный график был перекрашен в розовый при помощи параметра color

Шаг 3. Сделаем 3 штуки стобчатой диаграммы 

```python
plt.figure(figsize=(25, 10))
plt.bar(starWars['name'].head(11), starWars['height'].head(11), label='Рост', color = "orange")
plt.xlabel('Имя персонажа')
plt.ylabel('Рост')
plt.title('Рост персонажей')
plt. legend( )
plt.show()
```

Была сделана столбчатая диаграмма на которой отображен рост и имя первых десяти персонажей (например: рост Люка Скайуокера равен 172, рост C-3PO равен 167 и т.д. )

```python
plt.figure(figsize=(25, 10))
plt.bar(starWars['name'].head(11), starWars['mass'].head(11), label='Вес', color = "green")
plt.bar(starWars['name'].head(11), starWars['birth_year'].head(11), label='возраст', color = "red")
plt.xlabel('Имя персонажа')
plt.title('Вес и возраст персонажей персонажей')
plt. legend( )
plt.show()
```

Была сделана объединенная столбчатая диаграмма которая показывает вес и возраст первых одиннадцати пресонажей (Возраст-красным, вес - зеленым) (на диаграмме видно, что вес Люка Скайуокера равен 77, а возраст равен 19 и т.д. )

```python
plt.figure(figsize=(25, 10))
plt.bar(starWars['name'].tail(11), starWars['height'].tail(11), label='Рост', color = "yellow")
plt.xlabel('Имя персонажа')
plt.ylabel('Рост')
plt.title('Рост персонажей персонажей')
plt. legend( )
plt.show()
```

Была сделана столбчатая диаграмма которая отображает рост последних одиннадцати персонажей на диаграмме видно, что рост (Падме Амидалы равен 165 и т.д.)

Шаг 4. Сделаем 2 штуки круговых диаграмм 

```python
plt.figure(figsize=(25, 10))
Cvet_volos = starWars['hair_color'].value_counts()
plt.pie(Cvet_volos, labels= Cvet_volos.index, autopct='%1.lf%%',
colors=['purple', 'blue', 'orange', 'pink', 'red', 'green', 'grey', 'black', 'White', 'gold', 'brown', 'cyan'])
plt.title('Соотношение цвета волос персонажей')
plt.show()
```

Была сделана круговая диаграмма в результате которой видно, что у 45% персонажей нету волос, у 22% персонажей коричневые волосы, у 16% персонажей черные волосы, у 5% пресонажей белые и т.д.

```python
plt.figure(figsize=(25, 10))
Cvet_skin = starWars['skin_color'].value_counts()
plt.pie(Cvet_skin, labels= Cvet_skin.index, autopct='%1.lf%%',
colors=['purple', 'blue', 'orange', 'pink', 'red', 'green', 'grey', 'black', 'White', 'gold', 'brown', 'cyan'])
plt.title('Соотношение цвета кожи персонажей')
plt.show()
```

Была сделана круговая диаграмма в результате которой видно, что у 20% персонажей светлая кожа у 13% светлая кожа, у 7% темная кожа и т.д.

Шаг 5. Сделаем  2 диаграммы рассеяния

Шаг 6. Сделаем 3 гистограммы

Была создана гистограмма на которой изображено распределение роста пресонажей на ней видно, что самый большой рост у 14 персонажей а менбше всего персонажей у которых рост равен 75, 90, 110, 120, 133, 220, 230 и 260

Была создана гистограмма на которой видно что больше всего персонажей с сине -желтыми галазами (30), а меньше всего персонажей с зелеными и желтыми глазами (3)

Была сделана гистограмма на которой видео что больше всего персонажей с весом в диапозоне от 30 до 190 (>50), а меньше всего в диапозоне от 1250 до 1350
